```python
import seaborn as sns
%matplotlib inline
```

```python
tips = sns.load_dataset('tips')
```

```python
tips.head()
```

```python
sns.distplot(tips['total_bill'], bins= 40)
```

```python
sns.jointplot(x= 'total_bill',y= 'tip',data= tips, kind= 'hex')
```

```python
sns.pairplot(tips, hue= 'sex', palette= 'cool')
```

```python
sns.rugplot(tips['total_bill'])
```
