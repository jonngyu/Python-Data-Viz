# Brief introduction on how to plot with Matplotlib

```python
import matplotlib.pyplot as plt
import numpy as np
%matplotlib inline
```

```python
x= np.linspace(0,5,11)
y= x **2
```

## Using the funtional method to plot

```python

plt.plot(x,y)
plt.xlabel('X Label')
plt.ylabel('Y Label')
plt.title('Title')
```

### Creating subplots

```python
plt.subplot(1,2,1)
plt.plot(x,y,'r')

plt.subplot(1,2,2)
plt.plot(y,x,'b')
```

## Using object oriented method to plot

```python
fig= plt.figure()

axes= fig.add_axes([0.1,0.1,0.8,0.8])

axes.plot(x,y)
axes.set_xlabel('X Label')
axes.set_ylabel('Y Label')
axes.set_title('Set Title')
```

### Creating subplots

```python
fig= plt.figure()

axes1= fig.add_axes([0.1,0.1,0.8,0.8])
axes2= fig.add_axes([0.2,0.5,0.4,0.3])

axes1.plot(x,y)
axes1.set_title('Larger Plot')

axes2.plot(y,x)
axes2.set_title('Smaller Plot')
```

### Creating subplots through iteration

```python
fig,axes= plt.subplots(nrows=1,ncols=2)

for current_ax in axes:
    current_ax.plot(x,y)
    
plt.tight_layout()
```

### Creating subplots through indexing

```python
fig,axes= plt.subplots(nrows=1,ncols=2)

axes[0].plot(x,y)
axes[0].set_title('First Plot')

axes[1].plot(y,x)
axes[1].set_title('Second Plot')

plt.tight_layout()
```
