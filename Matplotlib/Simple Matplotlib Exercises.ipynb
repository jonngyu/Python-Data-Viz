```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline

x = np.arange(0,100)
y = x*2
z = x**2
```

# Exercise 1

```python
fig = plt.figure()

ax = fig.add_axes([0,0,1,1])
ax.plot(x,y)
ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.set_title('Title')
```

# Exercise 2

```python
fig = plt.figure()

#Creating plots
ax1 = fig.add_axes([0,0,1,1])
ax2 = fig.add_axes([0.2,.5,.2,.2])

#Add simple colours
ax1.plot(x,y,color='red')
ax2.plot(x,y,color='red')
```

# Exercise 3

```python
fig = plt.figure()

ax = fig.add_axes([0,0,1,1])
ax2 = fig.add_axes([0.2,.5,.4,.4])

#Large Plot
ax.plot(x,z)
ax.set_xlabel('X')
ax.set_ylabel('Z')

#Insert plot
ax2.plot(x,y)
ax2.set_title('Zoom')
ax2.set_xlabel('X')
ax2.set_ylabel('Y')
ax2.set_xlim([20,22])
ax2.set_ylim([30,50])
```

# Exercise 4

```python
fig, axes = plt.subplots (1,2, figsize=(12,2))

axes[0].plot(x,y,color='blue', ls= '--', lw= 3)


axes[1].plot(x,z, color= 'red', ls= '--', lw=3)
```
