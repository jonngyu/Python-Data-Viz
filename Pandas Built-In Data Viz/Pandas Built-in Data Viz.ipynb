# Demonstrating the built in data viz funtions in Pandas

```python
import numpy as np
import pandas as pd
import seaborn as sns
%matplotlib inline
import os as os
```

```python
df1= pd.read_csv('df1', index_col=0)
df2= pd.read_csv('df2')
df3= pd.read_csv('df3')
```

```python
df1['A'].hist(bins= 30)
```

```python
df1['A'].plot.hist()
```

```python
df2.plot.area(alpha=0.4)
```

```python
df2.plot.bar(stacked= True)
```

```python
df1['A'].plot.hist(bins= 50)
```

```python
df1.plot.line(x= df1.index, y='B', figsize=(12,3), lw= 1)
```

```python
df1.plot.scatter(x= 'A', y= 'B', s= df1['C']*50)
```

```python
df2.plot.box()
```
